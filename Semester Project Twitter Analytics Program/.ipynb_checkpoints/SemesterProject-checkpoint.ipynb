```python
# This loads in my files as pandas so I can filter the data as something I can mess with
TwitterFollowCount = pd.read_csv('Twitter Data - Sheet1.csv')
TwitterData_1 = pd.read_csv('twitonomy_trev_barry_trev_barry.csv - twitonomy_trev_barry_trev_barry.csv.csv')
TwitterData_1.head()
```

```python
# This is a seperate file that contains the number of followers I have at the start of each month
TwitterFollowCount.tail()
```

```python
# This displays all the tweets I have done including retweets and replies that I've done. For this project I'm only
# concerning myself with original tweets that I have created myself. To do this I will use a mask that filters out
# all the tweets that don't fall under the 'Unnamed: 6' category.

TwitterDataFiltered = TwitterData_1[TwitterData_1['Unnamed: 6']=='New']

# Need to create a new dataframe that dropps of the the columns that are not needed and converts the last two rows
# into floats from strings
TwitterDataFiltered = TwitterDataFiltered[['Unnamed: 0','Unnamed: 3','Unnamed: 7','Unnamed: 8']]
# I use the above pd file to read in using loadtxt for just the last two columns so I can read them in as floats
# and plot them
TwitterDataFiltered.head(n=10)
```

**Below is where I will be converting all of my data into formats that I can read into graphs and plot**

```python
# Creating a list of all of my likes and converting into floats
likes = []
for i in TwitterDataFiltered['Unnamed: 8']:
    likes.append(float(i))
```

```python
# This is to store the tweet length for each tweet
length = []
for i in range(0,len(TwitterDataFiltered['Unnamed: 3'])):
    length.append(len(TwitterDataFiltered['Unnamed: 3'].values[i]))
```

```python
# This is to create a list of my retweets
retweets = []
for i in TwitterDataFiltered['Unnamed: 7']:
    retweets.append(float(i))
```

```python
# This is to create a check to see if media was used and create a list from this
media = []
for i in range(0,len(TwitterDataFiltered['Unnamed: 3'].values)):
    string = TwitterDataFiltered['Unnamed: 3'].values[i]
    substring = 't.co'
    count = 0

    if string.find(substring) is not -1:
        count+=1
        media.append(count)
    else:
        count+=0
        media.append(count)
```

```python
# This is a check to see if tagging was used and create a list from this
tagging = []
for i in range(0,len(TwitterDataFiltered['Unnamed: 3'].values)):
    string = TwitterDataFiltered['Unnamed: 3'].values[i]
    substring = '@'
    count = 0

    if string.find(substring) is not -1:
        count+=1
        tagging.append(count)
    else:
        count+=0
        tagging.append(count)
```

**Below is where I will be plotting all of my information.**

```python
# For this regression model I will be taking into account 4 factors: # of Retweets, media usage, tweet length
# tagging or not

plt.scatter(TwitterFollowCount['Date Number'],TwitterFollowCount['Num of followers'],color='g')
plt.title('Number of Followers over Time')
plt.ylabel('Number of Followers')
plt.xlabel('Number of Months Since March 2018')

p = np.polyfit(TwitterFollowCount['Date Number'],TwitterFollowCount['Num of followers'],1)
my_func = np.poly1d(p)
plt.plot(TwitterFollowCount['Date Number'],my_func(TwitterFollowCount['Date Number']))
print('Number of followers=',my_func)
print('This tells me that on average I gain about 3.5 followers a month.')
plt.grid()
plt.show()
```

**The equation represents a function of number of likes a tweet gets based on the following parameters:**

**This is where I will be pulling everything from my above data into a regression model and finding all the coefficients to see what affected the likes on my tweets the most.**

```python
# Take in each variable and pull a regression line from it that determines the correlation between each data set
# when compared to the number of likes. From each data set it will return a linear function plus coeffiecents that
# will show how much impact each variable will have on my total number of likes
likes.reverse()
# Number of retweets
a = np.polyfit(retweets,likes,1)
my_func_retweets = np.poly1d(a)

# Media usage
b = np.polyfit(media,likes,1)
my_func_media = np.poly1d(b)

# Length of tweet
c = np.polyfit(length,likes,1)
my_func_length = np.poly1d(c)

# tagging or not
d = np.polyfit(tagging,likes,1)
my_func_tagging = np.poly1d(d)

# number of followers
# This is a seperate function that I will compare to the other function. I will take the sum of the total likes per
# month and sum them up to compare this to the total number of followers I had for that month
TwitterDataFiltered['Unnamed: 0'].head()

print('My function that defines my retweets is:',my_func_retweets)
print('My function that defines my media usage is:',my_func_media)
print('My function that defines the length of tweet is:',my_func_length)
print('My function that defines my use of tagging is:',my_func_tagging)
```

**I combine all of these above functions together to create my model of likes. Each function has an associated coefficient with it plus a 'b' value. The coefficients are attached to the variable while the 'b' value describes a "popularity index", meaning if all my variables in my equation were zero, I would still get around 12 base likes.**

L(t) = {1.614}{r(t)}+{0.4493}{m(t)}+{0.05002}{l(t)}-{1.789}{g(t)}+{11.966}
\end{equation}
